```python
import pandas as pd
import numpy as np
import tensorflowjs as tfjs
from keras import models
from keras import layers
```

```python
data_train = pd.read_csv(
    "train.csv")
data_train.head(5)
```

```python
data_test = pd.read_csv("test.csv")
data_test.head(5)
```

```python
def omitted_values(data):
    print(pd.isnull(data).sum()) # Verificamos que no tenga valores omitidos
```

```python
omitted_values(data_test)
```

```python
omitted_values(data_train)
```

```python
def get_input(data, test = False):
    if test:
        return data.drop(["ID"], axis = 1) 
    else:
        return data.drop(["ID", "TARGET"], axis = 1)
```

```python
def get_output(data):
    return data[["TARGET"]]
```

```python
input_train = get_input(data_train)
output_train = get_output(data_train)
```

```python
input_test = get_input(data_test, test = True)
```

```python
input_train.shape
```

```python
print(output_train.shape)
print(output_train.dtypes)
```

```python
input_test.shape
```

```python
mean_test = input_test.mean(axis = 0)
mean_test.to_csv("mean_test.csv", header=["value"], index = 1)
```

```python
def get_model(arr):
    model = models.Sequential()
    for i in range(len(arr)):
        if i!=0 and i!=len(arr)-1:
            if i==1:
                model.add(layers.Dense(arr[i],input_dim=arr[0],kernel_initializer='normal', activation='relu'))
            else:
                model.add(layers.Dense(arr[i],activation='relu'))
    model.add(layers.Dense(arr[-1],kernel_initializer='normal',activation="sigmoid"))
    model.compile(loss="binary_crossentropy",optimizer='rmsprop',metrics=['accuracy'])
    return model
```

```python
network = get_model([369, 200, 50, 1])
```

```python
network.fit(np.array(input_train), np.array(output_train), epochs = 20)
```

```python
preds = network.predict(np.array(input_test))
```

```python
tfjs.converters.save_keras_model(network, "user_satisfaction_network")
```
